```python
def findBracelet(x, y, a, all):
    print("START")
    
    
    a[0] = [x, y]
    all[int(a[0][0])][int(a[0][1])] = 0

    a[1] = [a[0][1], (a[0][0] + a[0][1]) % 10]
    all[int(a[1][0])][int(a[1][1])] = 0

    
    i = 1
    print(i, a[0])
    while (a[i][0] != a[0][0]) | ((a[i][1] != a[0][1])):
        i+=1
        a[i] = [a[i-1][1], (a[i-1][0] + a[i-1][1]) % 10]
        all[int(a[i][0])][int(a[i][1])] = 0
        print(i, a[i-1])

    
    print("END")
    print()
    print()
    print()
    print()
    print()
    
    
    return a, all
    
```

```python
grid = np.ones(shape=(10,10))

pairs = np.zeros(shape=(101, 2))


for i in range(0, 10):
    for j in range(0, 10):
        if grid[i][j] != 0:
            pairs, grid = findBracelet(i, j, pairs, grid)



```
